### Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load Data

```python
df=pd.read_csv('../recommendation/data.csv')
```

```python
df.describe().transpose()
```

```python
def get_artist_music(start=0, end=100):
    try:
        filtered_df = df[start:end]
        result = {
            'artists': filtered_df['artists'].to_list(),
            'name': filtered_df['name'].to_list(),
            'release_date': filtered_df['release_date'].to_list(),
            'year': filtered_df['year'].to_list(),
            'popularity': filtered_df['popularity'].to_list()
        
        }
    except Exception as e:
        print(e)
    
    return result
    
```

```python
print(get_artist_music(start=0, end=5))
```

### Music dictionary

```python
def get_music_features(start=0, end=1000):
    
    try:
        filtered_df = df[start:end]
        result = {
        'id': filtered_df['id'].to_list(),
        'name': filtered_df['name'].to_list(),
        'acousticness': filtered_df['acousticness'].to_list(),
        'danceability': filtered_df['danceability'].to_list(),
        'duration_ms': filtered_df['duration_ms'].to_list(),
        'energy': filtered_df['energy'].to_list(),
        'explicit': filtered_df['explicit'].to_list(),
        'instrumentalness': filtered_df['instrumentalness'].to_list(),
        'key': filtered_df['key'].to_list(),
        'liveness': filtered_df['liveness'].to_list(),
        'loudness': filtered_df['loudness'].to_list(),
        'mode': filtered_df['mode'].to_list(),
        'popularity': filtered_df['popularity'].to_list(),
        'speechiness': filtered_df['speechiness'].to_list(),
        'tempo': filtered_df['tempo'].to_list(),
        'valence': filtered_df['valence'].to_list()
        
    }
    except Exception as e:
        print(e)
    return result
```

```python
print(get_music_features(start=0, end=5))
```
