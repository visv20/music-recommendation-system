### Similarity Module

#### 1. Cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity, pairwise_distances
from scipy.spatial.distance import pdist,squareform
```

```python
def get_music_cosine_similarity(music_feature_dict, music_id_1, music_id_2):
    position1 = music_feature_dict['id'].index(music_id_1)
    position2 = music_feature_dict['id'].index(music_id_2)
    
    cosine_df = pd.DataFrame(data={
        'acousticness': [
            music_feature_dict['acousticness'][position1],
            music_feature_dict['acousticness'][position2]
        ],
        'danceability': [
            music_feature_dict['danceability'][position1],
            music_feature_dict['danceability'][position2]
        ],
        'energy': [
            music_feature_dict['energy'][position1],
            music_feature_dict['energy'][position2]
        ],
        'explicit': [
            music_feature_dict['explicit'][position1],
            music_feature_dict['explicit'][position2]
        ],
        'instrumentalness': [
            music_feature_dict['instrumentalness'][position1],
            music_feature_dict['instrumentalness'][position2]
        ],
        'key': [
            music_feature_dict['key'][position1],
            music_feature_dict['key'][position2]
        ],
        'liveness': [
            music_feature_dict['liveness'][position1],
            music_feature_dict['liveness'][position2]
        ],
        'loudness': [
            music_feature_dict['loudness'][position1],
            music_feature_dict['loudness'][position2]
        ],
        'mode':[
            music_feature_dict['mode'][position1],
            music_feature_dict['mode'][position2]
        ],
        'speechiness': [
            music_feature_dict['speechiness'][position1],
            music_feature_dict['speechiness'][position2]
        ],
        'tempo': [
            music_feature_dict['tempo'][position1],
            music_feature_dict['tempo'][position2]
        ],
        'valence': [
            music_feature_dict['valence'][position1],
            music_feature_dict['valence'][position2]
        ]
        
    })
    return cosine_similarity(cosine_df)
```

```python
music = get_music_features(start=0, end=3)
print('Music features cosine similarity')
print('--------------------------------')
get_music_cosine_similarity(music_feature_dict=music, music_id_1='6KbQ3uYMLKb5jDxLF7wYDD', music_id_2='6KuQTIu1KoTTkLXKrwlLPV')
```

#### 2. Jaccard similarity

```python
def get_music_jaccard_similarity(music_feature_dict, music_id_1, music_id_2):
    position1 = music_feature_dict['id'].index(music_id_1)
    position2 = music_feature_dict['id'].index(music_id_2)
    
    jaccard_df = pd.DataFrame(data={
        'acousticness': [
            music_feature_dict['acousticness'][position1],
            music_feature_dict['acousticness'][position2]
        ],
        'danceability': [
            music_feature_dict['danceability'][position1],
            music_feature_dict['danceability'][position2]
        ],
        'energy': [
            music_feature_dict['energy'][position1],
            music_feature_dict['energy'][position2]
        ],
        'explicit': [
            music_feature_dict['explicit'][position1],
            music_feature_dict['explicit'][position2]
        ],
        'instrumentalness': [
            music_feature_dict['instrumentalness'][position1],
            music_feature_dict['instrumentalness'][position2]
        ],
        'key': [
            music_feature_dict['key'][position1],
            music_feature_dict['key'][position2]
        ],
        'liveness': [
            music_feature_dict['liveness'][position1],
            music_feature_dict['liveness'][position2]
        ],
        'loudness': [
            music_feature_dict['loudness'][position1],
            music_feature_dict['loudness'][position2]
        ],
        'mode':[
            music_feature_dict['mode'][position1],
            music_feature_dict['mode'][position2]
        ],
        'speechiness': [
            music_feature_dict['speechiness'][position1],
            music_feature_dict['speechiness'][position2]
        ],
        'tempo': [
            music_feature_dict['tempo'][position1],
            music_feature_dict['tempo'][position2]
        ],
        'valence': [
            music_feature_dict['valence'][position1],
            music_feature_dict['valence'][position2]
        ]
        
    })
    jac_sim = 1 - pairwise_distances(jaccard_df.T, metric = "hamming")
    return jac_sim
    
```

```python
music = get_music_features(start=0, end=3)
print('Music features jaccard similarity')
print('---------------------------------')
get_music_jaccard_similarity(music_feature_dict=music, music_id_1='6KbQ3uYMLKb5jDxLF7wYDD', music_id_2='6KuQTIu1KoTTkLXKrwlLPV')
```

#### 3. Euclidean similarity

```python
def get_music_euclidean_similarity(music_feature_dict, music_id_1, music_id_2):
    position1 = music_feature_dict['id'].index(music_id_1)
    position2 = music_feature_dict['id'].index(music_id_2)
    
    euclidean_df = pd.DataFrame(data={
        'acousticness': [
            music_feature_dict['acousticness'][position1],
            music_feature_dict['acousticness'][position2]
        ],
        'danceability': [
            music_feature_dict['danceability'][position1],
            music_feature_dict['danceability'][position2]
        ],
        'energy': [
            music_feature_dict['energy'][position1],
            music_feature_dict['energy'][position2]
        ],
        'explicit': [
            music_feature_dict['explicit'][position1],
            music_feature_dict['explicit'][position2]
        ],
        'instrumentalness': [
            music_feature_dict['instrumentalness'][position1],
            music_feature_dict['instrumentalness'][position2]
        ],
        'key': [
            music_feature_dict['key'][position1],
            music_feature_dict['key'][position2]
        ],
        'liveness': [
            music_feature_dict['liveness'][position1],
            music_feature_dict['liveness'][position2]
        ],
        'loudness': [
            music_feature_dict['loudness'][position1],
            music_feature_dict['loudness'][position2]
        ],
        'mode':[
            music_feature_dict['mode'][position1],
            music_feature_dict['mode'][position2]
        ],
        'speechiness': [
            music_feature_dict['speechiness'][position1],
            music_feature_dict['speechiness'][position2]
        ],
        'tempo': [
            music_feature_dict['tempo'][position1],
            music_feature_dict['tempo'][position2]
        ],
        'valence': [
            music_feature_dict['valence'][position1],
            music_feature_dict['valence'][position2]
        ]
        
    })
    distances = pdist(euclidean_df, metric='euclidean')
    dist_matrix = squareform(distances)
    return dist_matrix
    
    
```

```python
music = get_music_features(start=0, end=3)
print('Music features euclidean similarity')
print('---------------------------------')
get_music_euclidean_similarity(music_feature_dict=music, 
                               music_id_1='6KbQ3uYMLKb5jDxLF7wYDD', music_id_2='6KuQTIu1KoTTkLXKrwlLPV')
```
